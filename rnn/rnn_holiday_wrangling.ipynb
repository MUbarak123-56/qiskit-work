```python
holiday_new = holiday[(holiday["type"] != "Work Day")]
holiday_new = holiday_new[(holiday_new["transferred"] != True)]
holiday_new
```

```python
holiday_new["type"].value_counts()
```

```python
holiday_new[holiday_new["date"] == "2014-06-25"]
```

```python
holiday_new = holiday_new.reset_index(drop=True)
```

```python
holiday_use = holiday_new[["date", "type", "locale"]]
```

```python
holiday_use.head()
```

```python
holiday_type = holiday_use[["date", "type"]].groupby(["date","type"], as_index = False).size().pivot(index = "date", columns = "type", values = "size").fillna(0)
holiday_type["date"] = holiday_type.index
holiday_type = holiday_type.reset_index(drop=True)
holiday_type.head()
```

```python
holiday_type[holiday_type["date"] == "2014-06-25"]
```

```python
holiday_locale = holiday_use[["date", "locale"]].groupby(["date","locale"], as_index = False).size().pivot(index = "date", columns = "locale", values = "size").fillna(0)
holiday_locale["date"] = holiday_locale.index
holiday_locale = holiday_locale.reset_index(drop=True)
holiday_locale.head()
```

```python
len(holiday_locale), len(holiday_type)
```

```python
meaningful_holidays = pd.merge(holiday_type, holiday_locale, on = "date", how = "inner")
meaningful_holidays.head()
```

```python
meaningful_holidays[meaningful_holidays["date"] == "2014-06-25"]
```

```python
holiday[holiday["date"] == "2014-06-25"]
```

```python
holiday_new["new"] = holiday_new["type"] + "_" + holiday_new["locale"]
holiday_new["new"].value_counts()
```

```python
holiday_combo = holiday_new[["date", "new"]].groupby(["date","new"], as_index = False).size().pivot(index = "date", columns = "new", values = "size").fillna(0)
holiday_combo["date"] = holiday_combo.index
holiday_combo = holiday_combo.reset_index(drop=True)
holiday_combo.head()
```

```python
holiday_combo.shape
```

```python
holiday_combo[holiday_combo["date"] == "2014-06-25"]
```

```python
holiday_combo.to_csv("holidays_updated.csv", index = False)
```
